```python
import pandas as pd
import datetime

```

```python
df_bova=pd.read_csv("G:\Trading\K11\Dados Históricos\historical_bova.csv",';')
df_ind=pd.read_csv("G:\Trading\K11\Dados Históricos\historical_ind.csv",';')
df=df_bova.merge(df_ind,on='Dates')
df['Dates']=df['Dates'].apply(lambda x: datetime.datetime.strptime(x,"%d/%m/%Y %H:%M"))
df['trade']=df['Dates'].apply(lambda x: x.minute)
df['5min']=df['trade'].apply(lambda x: x if x%5==0 else -1)
display(df)
df[df['5min']!=-1].to_excel('trades.xlsx')
```
