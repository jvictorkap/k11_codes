```python

def create_str_serie(dte_ref, dte_venc, pc, pv, qtd):
    dte_ref = pd.to_datetime(str(dte_ref))
    dte_venc = pd.to_datetime(str(dte_venc))
    dte_ref = dte_ref.strftime("%d%m%Y")
    dte_venc = dte_venc.strftime("%d%m%Y")
#     pc = pc / qtd
#     pv = pv / qtd
    return "-".join(
        [dte_ref, dte_venc, "{:.6f}".format(pv), "{:.6f}".format(pc)]
    ).replace(".", ",")
```

```python
def create_str_serie(row):
    expiration_date = row["Inst. Expiry Date"]
    strike = format_number(row["Strike Price"])
    call_or_put = row["Inst. Style"]
    month_letter = months[expiration_date.month - 1]
    year = expiration_date.strftime("%y")
    day = expiration_date.strftime("%d")

    if row["Inst. Model"] == "SPT":
        return row["Inst. Primary Asset Id"] + " BZ EQUITY"

    if row["Inst. Type"] == "LIS":
        return row["Inst. Primary Listed Id"]

    if row["Inst. Model"] == "VAN":
        return f"{month_letter}{year}{day}-{call_or_put}-{strike}".replace(".", ",")

    else:
        strike_barrier = format_number(row["Barrier Strike"])
        # rebate = ("%.4f" % row["Barrier Rebate"]).replace(".", ",")
        rebate = format_number(row["Barrier Rebate"])
        sulfix = "CONT" if "C" == row["Inst. Type"][0] else "DISC"
        barrier_type = to_barrier_type(row["Inst. Type"])
        return f"{month_letter}{year}{day}-{call_or_put}-{strike}-{barrier_type}-{strike_barrier}-{rebate}-{sulfix}".replace(
            ".", ","
        )

def format_number(n):
    n = str(round(n, 10))
    n = re.sub(r"\.?0+$", "", n)
    return n
def to_barrier_type(s):
    try:
        return "K" + s[2] + s[1]
    except Exception:
        return "KOD"
```
