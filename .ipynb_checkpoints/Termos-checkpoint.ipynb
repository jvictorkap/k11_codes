```python
def find_data(ticker, field):
    conn_mongo = pymongo.MongoClient(
        "mongodb://kapitalo11:kapitalo11@192.168.254.20:27017/?authSource=k11"
    )
    coll = conn_mongo["k11"]["bbgrt"]

    result = coll.find_one({"_id": ticker}, {field: 1})
    try:
        return result[field]
    except Exception:
        return 0

def create_key(row):
    return "BVMF|FUTURE|DI1|" + workdays.workday(
        row["mat"], days=2, holidays=workdays.load_holidays()
    ).strftime("%b%y")

def create_bbg_symbol(ticker, rep, sulf, asst):
    return ticker.replace(asst, rep) + f" {sulf}"
def get_contracts(table, asst, day=None):
    if day is None:
        day = workdays.workday(datetime.date.today(), -1, workdays.load_holidays("B3"))
    conn = ps.connect(
        "host='PGKPTL01' dbname='Risk_DB' user='kapitalo11' password='kapitalo11'"
    )
    query = f"SELECT tckrsymb as symbol, xprtndt as mat FROM b3up2data.{table} where rptdt='{day}' and asst='{asst}'"
    df = pd.read_sql(query, con=conn)
    conn.close()
    return df
def p_c_brl_fixed():
    """
    Field: [taxa_termo,fator,taxa]
    """

    df_contract = pd.DataFrame()
    df_contract = df_contract.append(
        [
            ["F", "Jan"],
            ["G", "Feb"],
            ["H", "Mar"],
            ["J", "Apr"],
            ["K", "May"],
            ["M", "Jun"],
            ["N", "Jul"],
            ["Q", "Aug"],
            ["U", "Sep"],
            ["V", "Oct"],
            ["X", "Nov"],
            ["Z", "Dec"],
        ]
    )
    df_contract.columns = ["code", "contract"]
    query = (
        "SELECT str_codigo as codigo, str_codigoserie as serie, dbl_saquestomaturity as du FROM tbl_ativosbmf where dte_data='"
        + datetime.datetime.today().strftime("%Y-%m-%d")
        + "' and str_codigo='DI1'"
    )
    db_conn = ps.connect(
        "host='PGKPTL01' dbname='db_Teste' user='k_geral' password='Kapitalo@01'"
    )
    df_fut = pd.read_sql(query, con=db_conn)
    db_conn.close()
#     t_selic = selic()
#     next_copom = [d for d in copom if d > date.today()][0]

    df_fut["code"] = df_fut.serie.str.slice(0, 1)
    df_fut = df_fut.merge(df_contract, on="code")
    df_fut["contract"] = df_fut["contract"] + df_fut["serie"].str.slice(1, 3)
    df_fut = df_fut.drop(["code"], axis=1)

    df_fut["key"] = "BVMF|FUTURE|" + df_fut["codigo"] + "|" + df_fut["contract"]
    df_fut = df_fut.drop(["codigo", "contract"], axis=1)
    df_price = get_contracts("interest_rate_futurecontractsinstrumentfile", "DI1")

    df_price["ticker_bbg"] = df_price["symbol"].apply(
        lambda row: create_bbg_symbol(row, "OD", "Comdty", "DI1")
    )
    df_price["key"] = df_price.apply(lambda row: create_key(row), axis=1)
    df_price["last"] = df_price["ticker_bbg"].apply(lambda x: find_data(x, "price"))

    df_fut = df_fut.merge(df_price, on="key").sort_values(by="du")

    df_fut["current_price"] = df_fut["last"]

    df_fut["current_price"] = df_fut["current_price"] / 100
    df_fut["du"] = df_fut["du"] - 1
    df_fut["ajuste"] = 100000 / np.power(
        1 + df_fut["current_price"], df_fut["du"] / 252
    )

    df_fut = df_fut.join(df_fut.shift(1).add_suffix("_ant"))
    df_fut = df_fut.drop(["serie_ant", "current_price_ant"], axis=1)

    df_fut.loc[df_fut.ajuste_ant.isnull(), "ajuste_ant"] = 100000
    df_fut.loc[df_fut.du_ant.isnull(), "du_ant"] = 0

    df_fut["taxa_termo"] = (
        np.power(
            df_fut.ajuste_ant / df_fut.ajuste, 252 / (df_fut["du"] - df_fut["du_ant"])
        )
        - 1
    )
    df_fut["du"] = df_fut["du"].astype(int)
    df_fut["du_ant"] = df_fut["du_ant"].astype(int)

    df_curva_realtime = pd.DataFrame({"du": range(0, df_fut.du.max() + 1, 1)})

    df_curva_realtime = df_curva_realtime.merge(
        df_fut[["du", "taxa_termo"]], on="du", how="left"
    )
    df_curva_realtime = df_curva_realtime.bfill()
    # df_curva_realtime.loc[df_curva_realtime['du'] <= du_until_copom, 'taxa'] = t_selic / 100
    # df_curva_realtime.loc[df_curva_realtime['du'] <= du_until_copom, 'taxa_termo'] = t_selic / 100

    df_curva_realtime["fator"] = np.power(1 + df_curva_realtime["taxa_termo"], 1 / 252)

    df_curva_realtime.loc[0, "fator"] = 1
    df_curva_realtime["fator"] = df_curva_realtime["fator"].cumprod()
    df_curva_realtime["taxa"] = (
        np.power(df_curva_realtime.fator, 252 / df_curva_realtime["du"]) - 1
    )

    df_curva_realtime = df_curva_realtime.iloc[1:]
    df_curva_realtime = df_curva_realtime.set_index("du", drop=True)

    return df_curva_realtime

curva=p_c_brl_fixed()
```

```python
def curva_pre(x,df_curve=curva):
    return df_curve.loc[x,"taxa"]

curva_pre(18)
```

```python
import workdays
import numpy as np
holidays_br= workdays.load_holidays("BR")
query= f"SELECT dte_data, str_fundo, str_mesa, str_estrategia, str_corretora, str_mercado, str_titulo, dte_dataref, dte_vencimento, dbl_quantidade, dbl_taxa, dbl_preco, str_codigo, str_serie, str_origem \
FROM public.tbl_boletasrf where (dte_vencimento='{dt.strftime('%Y-%m-%d')}' or dte_data='{dt.strftime('%Y-%m-%d')}') and str_estrategia= 'Termo Antecipacao'  and dbl_quantidade <0 order by str_titulo ASC "

df=pd.read_sql(query,db_conn_test)

df['data_neg']=df.apply(lambda row: datetime.datetime.strptime(row['str_serie'][0:8],'%d%m%Y') if row['dbl_quantidade']<0 else row['dte_dataref'], axis=1)

df['dt_neg_f']=df['data_neg'].apply(lambda x: workdays.workday(x,-2,holidays_b3))

df['dt_neg_f']=pd.to_datetime(df['dt_neg_f']).apply(lambda x: x.date())

df['chave']=df.apply(lambda row: str(row['str_titulo'][:(len(row['str_titulo'])-1)])+str((row['dt_neg_f']-datetime.date(1900,1,1)).days+2)+str((row['dte_vencimento']-datetime.date(1900,1,1)).days+2),axis=1)


df['valor_liquidacao']=df['dbl_preco']*df['dbl_quantidade']

df=df[['chave','valor_liquidacao','str_serie']]


df=df.groupby("chave").agg({'valor_liquidacao':sum})

df=df.reset_index()
antecipa=df

display(antecipa)

# antecipa.to_excel("antecipa.xlsx")

query= f"SELECT dt_dataposicao as data, REPLACE(str_descr, ',', '') as codigo,dt_data_emissao as data_neg, dt_data_vcto as vencimento,sum(dbl_quantidade) as quantidade, sum(dbl_valor_mercado) as ValorMercado, (sum(dbl_valor_mercado) / sum(dbl_quantidade)) as PU,(SUM(dbl_pu_emissao * dbl_quantidade) / sum(dbl_quantidade)) as PUEmissao FROM tbl_posicao_bradesco where str_carteira='012979' and dt_dataposicao='{dt_1.strftime('%Y-%m-%d')}' and dbl_quantidade<>0 \
and str_nomegrupo='TERMO DE AÇÕES' \
GROUP BY dt_dataposicao, REPLACE(str_descr, ',', ''), dt_data_emissao, dt_data_vcto ORDER BY dt_data_emissao DESC"

df=pd.read_sql(query,db_conn_test)
display(df)
# df['str_serie']=df.apply(lambda row: create_str_serie(row['data_neg], row['vencimento'], pc, pv, qtd),axis=1)

df['chave']=df.apply(lambda row: str(row['codigo'])+str((row['data_neg']-datetime.date(1900,1,1)).days+2)+str((row['vencimento']-datetime.date(1900,1,1)).days+2),axis=1)


df['du']=df.apply(lambda row: workdays.networkdays(row['data_neg'],row['vencimento'],holidays=holidays_br),axis=1)
df['du_venc']=df.apply(lambda row: workdays.networkdays(dt,row['vencimento'],holidays_br)-1,axis=1)
df['taxa dia']=df.apply(lambda row: (((1/row['puemissao'])**(1/row['du'])) -1) ,axis=1) 
antecipa=antecipa.reset_index()

gerenga=pd.merge(df,antecipa,on='chave',how='left')


gerenga['valor_liquidacao']=gerenga['valor_liquidacao']*-1

df=gerenga
cdi=((1+11.65/100)**(1/252)-1)
cdi=round(cdi,8)

df=df.fillna(0)
df=df[df['quantidade']!=0]
df['pnl']=df.apply(lambda row:(row['valor_liquidacao' ] - (row['valormercado']*row['valor_liquidacao']/row['quantidade']) + row['valormercado']*row['taxa dia']*((row['quantidade']-row['valor_liquidacao'])/row['quantidade']) ) if row['valor_liquidacao']!=0 else row['valormercado']*row['taxa dia'],axis=1)
df['over cdi']= df.apply(lambda row:row['pnl']-(row['valormercado']*cdi),axis=1)


df['taxa pre']=df['du_venc'].apply(lambda x: curva_pre(int(x)) if x!=0 else 0 )
df['pu mercado']= 1 /((1+df['taxa pre'])**(df['du_venc']/252))

df['delta']=df['pu mercado']- df['pu']


df['res mtm']=df['delta']*(df['quantidade']-df['valor_liquidacao'])+ df['valor_liquidacao']-(df['valormercado']*df['valor_liquidacao']/df['quantidade'])
df=df[df['valor_liquidacao']!=0]
df['over mtm']=df['res mtm']- df['pu']*df['quantidade']*cdi
df
```
